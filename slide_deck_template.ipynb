## Investigation Overview

>  This presentation aims to get insights out of the FordGOBike to see what are the most interesting relationships between each variable. what are the facts regarding the duration time and how it's related to another variable such as days or hours or even the type of user?

## Dataset Overview

> The dataset  contains data about duration in sec and minutes. start and end station like id or the name , dates in hours , days , months. station geographic like longitude and latitude. bike_id and user_type.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('Cleaned_df')
df.head()
```

## (Visualization 1)

> The records shows that the avg minutes in September , August and July are higher than other months

```python
base_color = sb.color_palette()[0]
plt.figure(figsize=(20,10))
sb.boxplot(data=df, x='start_month', y='duration_minute', color=base_color);
plt.ylim(0,20)
plt.xticks(rotation=0);
plt.xlabel('Month');
plt.title('Month And avg. Duration time ( in minutes ) ')
plt.ylabel(' Avg. Trip Duration in Minute');
```

## (Visualization 2)

> The avg duration time for regular customars are higher in june and july, coming after that august and september. for subscribers almost all months are close to each either.

```python
plt.figure(figsize=(25,10))
sb.boxplot(data = df , x = 'user_type' ,y= df.duration_minute, hue='start_month' )
plt.ylim(0,50)
plt.title('User Type And avg. Duration time ( in minutes ) ');
plt.ylabel('avg. Duration time');
```

## (Visualization 3)
> The relationship shows that the people who start use the bikes system at 03:00 hour has the maximum avg duration for both types of people Customers and subscribers.


```python
plt.figure(figsize=(25,10))
sb.pointplot(data=df, x='start_hourofday', y= 'duration_minute'  , hue='user_type');
plt.xlabel('Hour Of Day');
plt.title('Hour of Day  for each type');
```
